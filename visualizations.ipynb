Import Libraries and Database

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('Datasets/Travel_And_Disease.csv')
%matplotlib inline
```

Create Reusable Linear Regression Function

```python
def simple_linear_regression(x, y):
    m, b = np.polyfit(x, y, 1)  # Linear regression: y = mx + b
    return m, b
```

Configure Plots

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 8), dpi=120)
```

Plot Total Number of Deaths for Ages 0 - 17 vs Airline Traffic

```python
young_total = data[data['Age_Group'] == '0-17 years']
x1 = young_total['Total_Deaths'].values
y1 = young_total['US_Airline_Traffic'].values

ax1.scatter(x1, y1, color = 'black', label = 'Data points')

m1, b1 = simple_linear_regression(x1, y1)
y1_pred = m1 * x1 + b1
ax1.plot(x1, y1_pred, color = 'red', label = 'Linear regression')

ax1.set_title('0-17 years', fontsize = 20)
ax1.set_xlabel('Total Deaths', fontsize = 16)
ax1.set_ylabel('Airline Traffic', fontsize = 16)
ax1.legend(fontsize = 12)
```

Plot Total Number of Deaths for Ages 18 - 64 vs Airline Traffic

```python
mid_total = data[data['Age_Group'] == '18-64 years']
x2 = mid_total['Total_Deaths'].values
y2 = mid_total['US_Airline_Traffic'].values

ax2.scatter(x2, y2, color = 'black', label='Data points')

m2, b2 = simple_linear_regression(x2, y2)
y2_pred = m2 * x2 + b2
ax2.plot(x2, y2_pred, color = 'red', label = 'Linear regression')

ax2.set_title('18 - 64 years', fontsize = 20)
ax2.set_xlabel('Total Deaths', fontsize = 16)
ax2.set_ylabel('Airline Traffic', fontsize = 16)
ax2.legend(fontsize = 12)
```
